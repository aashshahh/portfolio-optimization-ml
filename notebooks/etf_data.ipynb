```python
import pandas as pd
import yfinance as yf
```

```python
tickers = ["SPY","QQQ","TLT","GLD","EEM","IWM","VNQ","LQD"]
start = "2005-01-01"
end = "2024-12-31"
```

```python
data = yf.download(tickers,start=start,end=end,
    auto_adjust=True)
```

```python
prices = data["Close"]
prices.head()
```

```python
prices.to_csv("../data/market_data.csv")
print("data collected")
```
